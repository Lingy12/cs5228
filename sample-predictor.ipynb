```python
import pandas as pd
def round_to_nearest_50(n):
    remainder = n % 50
    if remainder < 25:
        return n - remainder
    else:
        return n + (50 - remainder)
    
output_df = pd.read_csv('./my_pred.csv', index_col=0)
output_df['predicted'] = output_df['predicted'].map(lambda x: round_to_nearest_50(x))
output_df.to_csv('my_pred_rounded.csv', index=False)
```
